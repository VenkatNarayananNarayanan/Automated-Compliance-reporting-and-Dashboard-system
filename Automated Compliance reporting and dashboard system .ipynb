```python
from datetime import datetime
```

```python
df['KYC_Last_Updated'] = pd.to_datetime(df['KYC_Last_Updated'])
df['Last_Payment_Date'] = pd.to_datetime(df['Last_Payment_Date'])
```

```python
df
```

```python
df.head()
```

```python
# Get todayâ€™s date
today = pd.to_datetime('2025-07-11')
```

```python
# Rule 1: KYC older than 12 months
df['KYC_Violation'] = (today - df['KYC_Last_Updated']).dt.days > 365

# Rule 2: Payment overdue > 60 days
df['Payment_Overdue'] = (today - df['Last_Payment_Date']).dt.days > 60

# Rule 3: Balance > Credit Limit
df['Overlimit_Violation'] = df['Current_Balance'] > df['Credit_Limit']

# Rule 4: Status = Delinquent or Watchlist
df['High_Risk_Status'] = df['Account_Status'].isin(['Delinquent', 'Watchlist'])

# Rule 5: Late fees > â‚¹2000
df['Late_Fee_Alert'] = df['Late_Fees_Accrued'] > 2000
```

```python
df['Total_Violations'] = (
    df['KYC_Violation'].astype(int) +
    df['Payment_Overdue'].astype(int) +
    df['Overlimit_Violation'].astype(int) +
    df['High_Risk_Status'].astype(int) +
    df['Late_Fee_Alert'].astype(int)
)
```
